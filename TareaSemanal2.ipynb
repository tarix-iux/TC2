### Hallar la transferencia $T=\frac{Vo}{Vi}$ en función de $ωo$ y $Q$.
![EJ1](images/TS2_1.png)

#### La transferencia queda definida entonces por:

$$
T(s) = -\frac{R_3}{R_1} \cdot \frac{ \frac{1}{C^2 R_3^2} }{ s^2 + s \cdot \frac{1}{C R_2} + \frac{1}{C^2 R_3^2} }
$$

#### Si expresamos como $T(s) = K \cdot f(Q, W_0)$:

$$
T(s) = K \cdot \frac{ W_0 }{ s^2 + s \cdot \frac{W_0}{Q} + W_0^2 }
$$

$$

#### Si expresamos como $T(s) = K \cdot f(Q, W_0)$:

$$
T(s) = K \cdot \frac{ s \cdot \frac{W_0}{Q} }{ s^2 + s \cdot \frac{W_0}{Q} + W_0^2 }
$$

$$
